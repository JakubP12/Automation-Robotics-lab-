```python
import ipywidgets as widgets
from IPython.display import display
from pulp import *


# Define slider for Chicken Price
ChickenSlider = widgets.FloatSlider(
    value=0.013,
    min=0,
    max=0.5,
    step=0.001,
    description='Chicken Price',
    continuous_update=False
)

# Define slider for Beef Price
BeefSlider = widgets.FloatSlider(
    value=0.008,
    min=0,
    max=0.5,
    step=0.001,
    description='Beef Price',
    continuous_update=False
)

# Define slider for Mutton Price
MuttonSlider = widgets.FloatSlider(
    value=0.010,
    min=0,
    max=0.5,
    step=0.001,
    description='Mutton Price',
    continuous_update=False
)

# Define slider for Rice Price
RiceSlider = widgets.FloatSlider(
    value=0.002,
    min=0,
    max=0.5,
    step=0.001,
    description='Rice Price',
    continuous_update=False
)

# Define slider for Wheat Price
WheatSlider = widgets.FloatSlider(
    value=0.005,
    min=0,
    max=0.5,
    step=0.001,
    description='Wheat Price',
    continuous_update=False
)

# Define slider for Gel Price
GelSlider = widgets.FloatSlider(
    value=0.001,
    min=0,
    max=0.5,
    step=0.001,
    description='Gel Price',
    continuous_update=False
)

# Define slider for Sardines Price
SardinesSlider = widgets.FloatSlider(
    value=0.015,
    min=0,
    max=0.5,
    step=0.001,
    description='Sardines Price',
    continuous_update=False
)


output = widgets.Output()

# Define the base nutritional values (these do not change with sliders)
proteinPercent = {    "CHICKEN": 0.100,    "BEEF": 0.200,    "MUTTON": 0.150,    "RICE": 0.000,    "WHEAT": 0.040,    "GEL": 0.000,    "SARDINES": 0.180}
fatPercent = {        "CHICKEN": 0.080,    "BEEF": 0.100,    "MUTTON": 0.110,    "RICE": 0.010,    "WHEAT": 0.010,    "GEL": 0.000,    "SARDINES": 0.050}
fibrePercent = {      "CHICKEN": 0.001,    "BEEF": 0.005,    "MUTTON": 0.003,    "RICE": 0.100,    "WHEAT": 0.150,    "GEL": 0.000,    "SARDINES": 0.000}
saltPercent = {       "CHICKEN": 0.002,    "BEEF": 0.005,    "MUTTON": 0.007,    "RICE": 0.002,    "WHEAT": 0.008,    "GEL": 0.000,    "SARDINES": 0.004}
vitBContent = {       "CHICKEN": 0.3,      "BEEF": 2.6,      "MUTTON": 2.6,      "RICE": 0.0,      "WHEAT": 0.0,      "GEL": 0.0,      "SARDINES": 8.9} # <<< Vitamin B Content Dictionary >>>

Ingredients = ["CHICKEN", "BEEF", "MUTTON", "RICE", "WHEAT", "GEL", "SARDINES"]

def solve_whiskas_problem(chicken_price, beef_price, mutton_price, rice_price, wheat_price, gel_price, sardines_price):
    # Update costs based on slider values
    costs = {
        "CHICKEN": chicken_price,
        "BEEF": beef_price,
        "MUTTON": mutton_price,
        "RICE": rice_price,
        "WHEAT": wheat_price,
        "GEL": gel_price,
        "SARDINES": sardines_price
    }

    # Create the 'prob' variable to contain the problem data
    prob = LpProblem("The Whiskas Problem", LpMinimize)

    # A dictionary called 'ingredient_vars' is created to contain the referenced Variables
    ingredient_vars = LpVariable.dicts("Ingr", Ingredients, lowBound=0)

    # The objective function is added to 'prob' first
    prob += (
        lpSum([costs[i] * ingredient_vars[i] for i in Ingredients]),
        "Total Cost of Ingredients per can",
    )

    # The five constraints are added to 'prob'
    prob += lpSum([ingredient_vars[i] for i in Ingredients]) == 100, "PercentagesSum"
    prob += (
        lpSum([proteinPercent[i] * ingredient_vars[i] for i in Ingredients]) >= 8.0,
        "ProteinRequirement",
    )
    prob += (
        lpSum([fatPercent[i] * ingredient_vars[i] for i in Ingredients]) >= 6.0,
        "FatRequirement",
    )
    prob += (
        lpSum([fibrePercent[i] * ingredient_vars[i] for i in Ingredients]) <= 2.0,
        "FibreRequirement",
    )
    prob += (
        lpSum([saltPercent[i] * ingredient_vars[i] for i in Ingredients]) <= 0.4,
        "SaltRequirement",
    )
    prob += lpSum([(vitBContent[i]) * ingredient_vars[i] for i in Ingredients]) >= 0.8, "VitaminBRequirement"

    # Solve the problem
    prob.solve()

    results = {
        "status": LpStatus[prob.status],
        "composition": {},
        "total_cost": value(prob.objective)
    }

    if LpStatus[prob.status] == 'Optimal':
        for i in Ingredients:
            results["composition"][i] = ingredient_vars[i].varValue
    return results


def update_whiskas_results(change):
    with output:
        output.clear_output()
        results = solve_whiskas_problem(
            ChickenSlider.value,
            BeefSlider.value,
            MuttonSlider.value,
            RiceSlider.value,
            WheatSlider.value,
            GelSlider.value,
            SardinesSlider.value
        )

        print("Status:", results['status'])
        if results['status'] == 'Optimal':
            print("--- Final Product Composition ---")
            for i, percentage in results['composition'].items():
                print(f"{i}: {percentage:.2f}%")
            print("---------------------------------")
            print(f"Total Cost: ${results['total_cost']:.4f}")
        else:
            print("Requirement cannot be met with current ingredients.")


# Connect sliders to the update function
ChickenSlider.observe(update_whiskas_results, names='value')
BeefSlider.observe(update_whiskas_results, names='value')
MuttonSlider.observe(update_whiskas_results, names='value')
RiceSlider.observe(update_whiskas_results, names='value')
WheatSlider.observe(update_whiskas_results, names='value')
GelSlider.observe(update_whiskas_results, names='value')
SardinesSlider.observe(update_whiskas_results, names='value')


# Display the sliders and initial results
display(ChickenSlider)
display(BeefSlider)
display(MuttonSlider)
display(RiceSlider)
display(WheatSlider)
display(GelSlider)
display(SardinesSlider)
display(output)

# Show initial results
update_whiskas_results(None)
```

Best blending solution currently relies on Beef and Gel due to their low cost-per-gram of protein. However, if Beef prices rise above $0.013, the solution shifts entirely to Chicken. The introduction of a Vitamin B requirement forces the inclusion of Sardines, which, despite their high cost, are the most efficient way to meet the nutritional floor.

In linear programming, the solution doesn't change gradually; it stays the same until a "shadow price" or a threshold is hit, then it jumps to a new ingredient mix
